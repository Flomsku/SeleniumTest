```python
from selenium import webdriver #Web driver import
from selenium.webdriver.common.keys import Keys #import key use
from selenium.webdriver.common.by import By #import searh id,tagname, class with By
from selenium.webdriver.support.ui import WebDriverWait #import waiting for loading
from selenium.webdriver.support import expected_conditions as EC #import for found conditions
import time

PATH = "C:\Program Files (x86)\chromedriver.exe" #Path where chromedrive.exe lays
driver = webdriver.Chrome(PATH)

driver.get("https://tilastokeskus.fi/hae") #get url

#print(driver.page_source) #print source code for the page

searchbar = driver.find_element(By.ID,"id_q") #looking for element by id_q and storing it to seachbar

searchbar.send_keys("inflaatio") #send keys to searchbar
searchbar.send_keys(Keys.RETURN) #hit enter

try:
    main = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "page")))
    articles = main.find_elements(By.TAG_NAME,"article")
    for article in articles:
        header = article.find_element(By.TAG_NAME,"h3")
        print(header.text)
        
finally:
    time.sleep(5)
#driver.close() 






#driver.quit() #close the browser 
#driver.close() #close tab
```

```python
from selenium import webdriver #Web driver import
from selenium.webdriver.common.keys import Keys #import key use
from selenium.webdriver.common.by import By #import searh id,tagname, class with By
from selenium.webdriver.support.ui import WebDriverWait #import waiting for loading
from selenium.webdriver.support import expected_conditions as EC #import for found conditions
import time

PATH = "C:\Program Files (x86)\chromedriver.exe" #Path where chromedrive.exe lays
driver = webdriver.Chrome(PATH)
driver.get("https://tilastokeskus.fi")
driver.find_element(By.PARTIAL_LINK_TEXT, 'Väki').click()
trs = driver.find_elements(By.TAG_NAME, 'tr')
for tr in trs:
    content = tr.find_element(By.TAG_NAME,"th")
    print(content.text)

```
